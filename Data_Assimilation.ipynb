For a system with a $D$ dimensional state variable $\vec{x}$ (for example, $D=4$ for a NaKL neuron), the conditional probability distribution $P$ of the system's path $𝐗$ on the measurement path $𝐘$, at time $t = n+1$, is:

$$P
\left(
𝐗\color{gray}{(n+1)}
| 
𝐘\color{gray}{(n+1)}
\right) 
= ℐ.ℳ.P(
𝐗\color{gray}{(n)}
|
𝐘\color{gray}{(n)}
)$$

where

𝐗: $\{
\vec{x}(t_0),\ 
\vec{x}(t_1 = t_0 + \Delta t),\ 
...,\ 
\vec{x}(t_{f} = t_0 + N\Delta t) 
\}$, and thus dimensions $D.(n+1)$,

𝐘: $\{
\vec{y}(\tau_0),\ 
\vec{y}(\tau_1),\ 
...,\ 
\vec{y}(\tau_{f} = t_0 + kN\Delta t) 
\}$, and thus dimensions $L.(n/k+1)$,

ℐ: represents information transfer, and

ℳ: represents the model.

Here, the subscripts denote time indices.

Notation:

$𝐗(n+1) = \{\vec{x}(n+1), 𝐗(n)\}$


$𝐘(n+1) = \{\vec{y}(n+1), 𝐘(n)\}$

$$\begin{align}
P\left(
𝐗\color{gray}{(n+1)} | 𝐘\color{gray}{(n+1)}\right) 
&= \frac{
P\left( 𝐗\color{gray}{(n+1)},\ 𝐘\color{gray}{(n+1)}\right)}
{P\left(𝐘\color{gray}{(n+1)}\right)} \\
&= \frac{
P\left(
\vec{x}\color{gray}{(n+1)},\ 𝐗\color{gray}{(n)},\ 
\vec{y}\color{gray}{(n+1)},\ 𝐘\color{gray}{(n)}\right)}
{P\left(
\vec{y}\color{gray}{(n+1)}, 𝐘\color{gray}{(n)}\right)} \\
&= \frac{
P\left(
\vec{x}\color{gray}{(n+1)},\ 𝐗\color{gray}{(n)},\ 
\vec{y}\color{gray}{(n+1)} \color{red}{ | 𝐘(n)}\right)}
{P\left(
\vec{y}\color{gray}{(n+1)} \color{green}{| 𝐘(n)}\right)}
. \frac{\color{green}{P(𝐘(n))}}{\color{red}{P(𝐘(n))}}\\
&= \frac{
P\left(
\vec{x}\color{gray}{(n+1)},\ 𝐗\color{gray}{(n)},\ 
\vec{y}\color{gray}{(n+1)} | 𝐘\color{gray}{(n)}\right)}
{P\left(
\vec{y}\color{gray}{(n+1)} | 𝐘\color{gray}{(n)}\right)}
. \color{orange}{
\frac
{P\left(\vec{x}(n+1) | 𝐗(n),\ 𝐘(n)\right)}
{P\left(\vec{x}(n+1) | 𝐗(n),\ 𝐘(n)\right)}
} \\
&= \frac{
P\left(
\vec{x}\color{gray}{(n+1)},\ 𝐗\color{gray}{(n)},\ 
\vec{y}\color{gray}{(n+1)} | 𝐘\color{gray}{(n)}\right)}
{P\left(
\vec{y}\color{gray}{(n+1)} | 𝐘\color{gray}{(n)}\right)}
. \frac
{P\left(\vec{x}\color{gray}{(n+1)} | 𝐗\color{gray}{(n)},\ 𝐘\color{gray}{(n)}\right)}
{
\color{purple}{
P\left(\vec{x}(n+1),\ 𝐗(n),\ 𝐘(n)\right)}
}
. \color{purple}{P\left(𝐗(n),\ 𝐘(n)\right)} \\
&= \frac{
P\left(
\vec{x}\color{gray}{(n+1)},\ 𝐗\color{gray}{(n)},\ 
\vec{y}\color{gray}{(n+1)} | 𝐘\color{gray}{(n)}\right)}
{P\left(
\vec{y}\color{gray}{(n+1)} | 𝐘\color{gray}{(n)}\right)}
. \frac
{P\left(\vec{x}\color{gray}{(n+1)} | 𝐗\color{gray}{(n)},\ 𝐘\color{gray}{(n)}\right)}
{
\color{purple}{
P\left(\vec{x}(n+1),\ 𝐗(n),\ 𝐘(n)\right)}
}
. \frac
{\color{purple}{P\left(𝐗(n),\ 𝐘(n)\right)}}
{\color{red}{P\left(𝐘(n)\right)}}
. \color{red}{P\left(𝐘(n)\right)} \\
&= \frac{
P\left(
\vec{x}\color{gray}{(n+1)},\ 𝐗\color{gray}{(n)},\ 
\vec{y}\color{gray}{(n+1)} | 𝐘\color{gray}{(n)}\right)}
{P\left(
\vec{y}\color{gray}{(n+1)} | 𝐘\color{gray}{(n)}\right)}
. \frac
{P\left(\vec{x}\color{gray}{(n+1)} | 𝐗\color{gray}{(n)},\ 𝐘\color{gray}{(n)}\right)}
{
\color{purple}{
P\left(\vec{x}(n+1),\ 𝐗(n),\ 𝐘(n)\right)}
}
.\color{purple}{P\left(𝐗(n) \color{red}{|  𝐘(n)}\right)}
. \color{red}{P\left(𝐘(n)\right)}
\end{align}$$
